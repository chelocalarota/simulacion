### Ejercicio 1.10

La cantidad de numeros validos total que se pueden formar con 3 bits es 8, entonces la 
Probabilidad de encontrar un numero de S es 6/8

```python
import numpy as np
def generateRN(n): 
    cant=3 
    x=[] 
    i=0    
    while i < n:                      
        coins=((np.random.rand(cant))>0.5) 
        decimal=binaryToInt(coins) 
        if decimal<=5: 
            x.append(decimal) 
            i=i+1     
    return x

N=10000
 
def binaryToInt(bitlist): 
    out = 0 
    for bit in bitlist: 
        out = (out << 1) | bit 
    return out
         
x=generateRN(N)
```

```python
import matplotlib.pyplot as plt 
plt.hist(x,bins=[1,2,3,4,5,6,7])
```
